```python
for slug in book_list_1['slug']:
    user_ids = []
    for page in range(1, 11):
        driver = webdriver.Chrome('/Users/connie/Downloads/chromedriver 2')
        driver.get(f'https://www.goodreads.com/book/show/{slug}?hide_last_page=true&page={i}')
        WebDriverWait.until(EC.presence_of_element_located(By.xpath("//div[@id='bookReviews']/div[@class='friendReviews elementListBrown'][1]/div[@class='section firstReview']//span[1]/a[@class='user']")))
        driver.implicitly_wait(10)
        source = driver.page_source
        driver.quit()
        soup = BeautifulSoup(source, 'lxml')
        book_reviews = soup.find("div", {'id':"bookReviews"})
        for i in book_reviews.find_all('a', {'class':"left imgcol"}):
            user_id = {}
            user_id['book_slug'] = slug
            user_id['user_slug'] = i['href']
            user_id['id'] = re.findall(r'\/user\/show\/(\d+)-\w+', i['href'])
            user_ids.append(user_id) 
    pd.DataFrame(user_ids).to_csv(f"./data/user_ids/{slug}")
    print(f'Finished gathering users for {slug}')
    time.sleep(1)
```
